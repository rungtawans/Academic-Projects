```python
# calculate mean likeCount each channel
from pymongo import aggregation
```

```python
results = videos.aggregate([
    {
    "$group": {
        "_id": "$channelId",
        "avg": { "$avg": "$likeCount"}
    }
}])
results = list(results)
results
```

```python
query = {"likeCount": {'$exists': False}}
ids = []
for item in list(videos.find(query)) :
  if item['channelId'] not in ids :
    ids.append(item['channelId'])
len(ids)
```

```python
query = { 'likeCount': { '$type' : 'null'} }
for d in results:
  if d['_id'] in ids:
    query = { "channelId" : d['_id']}
    print(list(videos.find(query)))
    new_value = { "$set" : {"likeCount": d['avg']}}
    videos.update_many(query, new_value)
```
