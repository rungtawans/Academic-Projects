```python
df_artist_spoftify = pd.read_csv('/content/artists-100row.csv')
df_artist_spoftify.drop(columns = ['Unnamed: 0'], inplace = True)
print(df_artist_spoftify.columns)
print(df_artist_spoftify.shape)
df_artist_spoftify
```

```python
f = open('/content/artist-100row-limit5.json')
data = json.load(f)

videoId = []
publishedAt = []
channelId = []
title = []
channelTitle = []
publishTime = []
description = []
for i in data:
    for j in i.get('items'):
      videoId.append(j.get('id').get('videoId'))
      publishedAt.append(j.get('snippet').get('publishedAt'))
      publishTime.append(j.get('snippet').get('publishTime'))
      title.append(j.get('snippet').get('title'))
      channelId.append(j.get('snippet').get('channelId'))
      channelTitle.append(j.get('snippet').get('channelTitle'))
      description.append(j.get('snippet').get('description'))
# Closing file
f.close()
```

```python
dict = {'videoId': videoId, 'publishedAt': publishedAt, 'publishTime': publishTime, 'title': title, 'channelId': channelId, 'channelTitle': channelTitle, 'description' : description}
df1 = pd.DataFrame(dict)
df1
```

```python
title_search = []
idx = []
for name in df_artist_spoftify['name'] :
  for title in df1['title'] :
    pattern = name
    res_findall = re.findall(pattern, title)
    if (len(res_findall) != 0):
      tempidx = df1.index[df1['title']==title]
      title_search.append(title)
      idx.append(tempidx[0])
      print(tempidx)
      print(title, res_findall)
      print("------------------")
```

```python
new_df = pd.DataFrame(df1, index = idx)
new_df

```

```python
new_df.drop_duplicates(subset=['channelId'], inplace =True)
print(new_df.shape)
new_df
```
